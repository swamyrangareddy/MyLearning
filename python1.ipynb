# **Strings - working with strings**

```python
#String Concatenation: Joining strings together is called string concatenation.
a = "Hello" + " " + "World"
print("string  concatenation :" ,a)

#String Repetition: * operator is used for repeating strings any number of times as required.
a = "@ " * 10
print("String Repetition :", a)

#Length of String: len() returns the number of characters in a given string.
user = input()
length = len(user)
print("Length of String :",length)

# String Indexing: We can access an individual character in a string using their positions (which start from 0) . These positions are also called index.
user = "Python"
first_letter = user[0]
print("String Indexing :", first_letter)

#String Slicing:  Obtaining a part of a string is called string slicing. Start from the start_index and stops at the end_index. (end_index is not included in the slice).
part = user[1:4]
print("String Slicing :",part)

#slicing to End: If end_index is not specified, slicing stops at the end of the string.
part = user[1:]
print("slicing to End :", part)

# Slicing from Start: If the start_index is not specified, the slicing starts from the index 0.
message = "Hi Steve"
part = message[:2]
print("Slicing from Start :",part)

#Reversing String: Reverse the given string using the extended slice operator.
txt = "Hello World"
print("Reversing String :",txt[::-1])
```

# **String Formatting**
String Formatting : String Formatting simplifies the concatenation. It increases the readability of code and type conversion is not required.

Add Placeholders : Add placeholders {} where the string needs to be formatted.

```python
name = "swamy"
age = 23
msg = "Hi {}. You are {} years old."
print(msg.format(name, age))

# Numbering Placeholders: Numbering placeholders, will fill values according to the position of arguments.

name = input("name :")
age = int(input("age :"))
msg = "Hi {1}. You are {0} years old."
print(msg.format(name, age))

# Naming Placeholder: Naming placeholders will fill values according to the keyword arguments.

name = input("name :")
age = int(input("age :"))
msg = "Hi {name}. You are {age} years old."
print(msg.format(age = age, name= name))
```

Every set element is

```python
# creating a set
a = 2
set_a = {5, "nine", a, 9.8,True}
print(type(set_a))
print(set_a)
```

# **Set Operations**
* Union
* Difference
* Intersection
* Symmetric Difference

* setunion.avif



```python
# Union of two sets is a set containing all elements of both sets.
# set_a | set_b or set_a.union(sequence)
# union() converts sequence to a set, and performs the union.
set_a = {4, 2, 8}
set_b = {1, 2}
union = set_a | set_b
print(union)
```

```python
# Intersection: The intersection of two sets is a set containing common elements of both sets.
# Syntax: set_a & set_b (or) set_a.intersection(sequence)
set_a ={ 1, 2, 3}
set_b ={4, 2, 6}
intersection = set_a & set_b
print(intersection)
```

```python
# Difference: The difference of two sets is a set containing all the elements in the first set but not the second.
# Syntax: set_a - set_b (or) set_a.difference(sequence)
set_a ={ 1, 2, 3}
set_b ={4, 2, 6}
diff = set_a - set_b
print(diff)
```

```python
# Symmetric Difference: Symmetric difference of two sets is a set containing all elements which are not common to both sets.
# Syntax: set_a ^ set_b (or) set_a.symmetric_difference(sequence)
set_a ={ 1, 2, 3}
set_b ={4, 2, 6}
symmetric_diff = set_a^ set_b
print(symmetric_diff)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

class Circle(object):

    # Constructor
    def __init__(self, radius=3, color='blue'):
        self.radius = radius
        self.color = color

    # Method
    def add_radius(self, r):
        self.radius = self.radius + r
        return(self.radius)

    # Method
    def drawCircle(self):
        plt.gca().add_patch(plt.Circle((0, 0), radius=self.radius, fc=self.color))
        plt.axis('scaled')
        plt.show()
```

```python
greenCircle = Circle(10,"green") # create an object
greenCircle.drawCircle() # draw the circle
```

```python
# Create a new Rectangle class for creating a rectangle object

class Rectangle(object):

    # Constructor
    def __init__(self, width=2, height=3, color='r'):
        self.height = height
        self.width = width
        self.color = color

    # Method
    def drawRectangle(self):
        plt.gca().add_patch(plt.Rectangle((0, 0), self.width, self.height ,fc=self.color))
        plt.axis('scaled')
        plt.show()
```

```python

pinkrectangle = Rectangle(4,5,"pink")
pinkrectangle.drawRectangle()
```

# **A Better Way to Open a File**
Using the with statement is better practice, it automatically closes the file even if the code encounters an exception. The code will run everything in the indent block then close the file object.

```python
# open file using with
with open(url,"r") as file1:
    FileContent = file1.read()
    print(FileContent)
```

```python
# verify if the file is closed
file1.closed
```

```python
# See the content of file
print(FileContent)
```

```python
# Read first four lines
with open(url,"r") as file1:
  print(file1.read())
```

```python
# read one line
with open(url,"r") as file1:
  print(file1.readline())
```

```python
with open(url,"r") as file1:
    print(file1.readline(20)) # does not read past the end of line
    print("----------")
    print(file1.read(20)) # Returns the next 20 chars
```

```python
# Iterate through the lines
with open (url,"r") as file1:
    i = 0;
    for line in file1:
        print("Iteration", str(i), ": ", line)
        i = i + 1
```

```python
with open(url,"r") as file1:
    filelist = file1.readlines()
    print("first line : ",filelist[0]) # print the first line
    print("last line : ",filelist[-1])
```
