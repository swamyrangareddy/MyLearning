# **Series**
A Pandas Series is like a column in a table.

It is a one-dimensional array holding data of any type.

```python
#  Create a simple Pandas Series from a list :
import pandas as pd

data = [3,6,8,9,3]
a = pd.Series(data)
print(a)
```

```python
# With the index argument, you can name your own labels.
import pandas as pd

data = [3,6,8,9,3]
a = pd.Series(data, index = ["a","b","c","d","e"])
print(a)
```

```python
# This label can be used to access a specified value.
print(a[0])

# When you have created labels, you can access an item by referring to the label.
print(a["b"])
```

```python
# we can also use arrays, when creating a Series
import numpy as np

data = np.array([10,20,30,40,50])
a = pd.Series(data)
print(a)
```

```python
# we can also use Tuples ,when creating a Series

data = [("a",10),("b",20),("c",30)]
a = pd.Series(data)
print(a)

# here complete tuple consider as a single element
```

```python
# You can also use a key/value object, like a dictionary, when creating a Series.

data = {"day1" : 120, "day2" : 150 , "day3" : 190}
ds = pd.Series(data)
print(a)
```

```python
# Create a Series using only data from "day1" and "day2":

data = {"day1" : 120, "day2" : 150 , "day3" : 190}
ds = pd.Series(data,index= ["day1","day2"])
print(ds)
```

# **DataFrame**
A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.

```python
# Create a simple Pandas DataFrame:
# Load a Python Dictionary into a DataFrame :
import pandas as pd

data = {
    "NAME" : ["steve","maccandless","sri sri"],
    "YEAR" : [1955,1968,1910],
    "BOOK" :["Steve jobs","into the wild","Maha Prasthanam"]
}
df = pd.DataFrame(data)
print(df)
```

```python
# Pandas use the loc attribute to return one or more specified row(s)
print(df.loc[0])
print("----------")

# #use a list of indexes:
print(df.loc[[0,2]])

# Note: When using [], the result is a Pandas DataFrame.
```

```python
# With the index argument, you can name your own indexes.
df = pd.DataFrame(data, index = ["first","second","third"])
print(df)
```

# **dataFrame Loc & iLoc**

**Loc :**
* Each iteration produces an index object and a row
* To access more than one row, use double brackets and specify the labels, separated by commas: df.loc[ [colum_1,column_2] ]
* Specify columns by including their labels in another list: df.loc[1,[column_3,column_6]]
* You can also specify a slice of the DataFrame with from and to labels, separated by a colon: df.loc[column_1, column_6]
* When slicing, both from and to are included in the result.

**ILoc :**
* The iloc property gets, or sets, the value(s) of the specified indexes.
* Specify both row and column with an index.
* To access more than one row, use double brackets and specify the indexes, separated by commas: df.iloc[[1,2]]
* Specify columns by including their indexes in another list:
df.iloc[[0, 2], [0, 1]]
* You can also specify a slice of the DataFrame with from and to indexes, separated by a colon: df.iloc[0:2]
* When slicing, the to index is excluded from the result.
